Checking for Outliers
*Multivariate Analysis using IQR Interquartile Range

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
IQR
```

```python
df.drop('id', axis = 1,  inplace = True)
```

```python
### Checking the actual datapoint that is an outlier
(df < (q1 -1.5 * IQR)) | (df >(q3 + 1.5 * IQR))
```

The TRUE show where there are Outliers 

```python
df.head()
```

```python
df_no_outlier = df[~((df < (q1 -1.5 * IQR)) | (df >(q3 + 1.5 * IQR))).any(axis=1)]
```

```python
df_no_outlier
```

```python
print(df.shape)
print(df_no_outlier.shape)
```

Feature Selection and Importance

+SELECTKBEST(Strong relation with the output/target)


```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection  import chi2
```

```python
df.columns
```

```python
df.rename(columns={'absenteeism_time_in_hours':'target_class'}, inplace = True)
```

```python
df.columns
```

```python
x = df[['reason_for_absence', 'month_of_absence', 'day_of_the_week', 'seasons',
       'transportation_expense', 'distance_from_residence_to_work',
       'service_time', 'age', 'work_load_average/day_', 'hit_target',
       'disciplinary_failure', 'education', 'son', 'social_drinker',
       'social_smoker', 'pet', 'weight', 'height', 'body_mass_index']]
y = df['target_class']
```

```python
# using SKBEST
skb = SelectKBest(score_func=chi2,k=10)
best_feature_fit = skb.fit(x,y)
```

```python
# Highest score shows the best fit
print("Scores",best_feature_fit.scores_)
```

```python
# Mapping to Feature Name
feature_scores = pd.DataFrame(best_feature_fit.scores_,columns=['Feature_Scores'])
```

```python
feature_scores
```

```python
feature_column_names = pd.DataFrame(x.columns,columns=['Feature_name'])
best_feat_df = pd.concat([feature_scores,feature_column_names],axis=1)
```

```python
best_feat_df
```

```python
# Get the best
# The higher the number the more important the feature
best_feat_df.nlargest(12,'Feature_Scores')
```

```python
best_feat_df.nlargest(12,'Feature_Scores')['Feature_name'].unique()
```

MODEL BUILDING 

```python
# ML Pkg
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn import svm 

# Metrics
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# Features for Initial Data
df.columns
```

```python
x
```

```python
y = df['target_class']
```

```python
df.columns
```

```python
x_best = df[['reason_for_absence','day_of_the_week','distance_from_residence_to_work','service_time','age','work_load_average/day_','disciplinary_failure','son','social_drinker','pet','height', 'body_mass_index']]
```

```python
# train /test dataset general 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=101)
```

```python
# train /test dataset best X  
x_train2,x_test2,y_train2,y_test2 = train_test_split(x_best,y,test_size=0.25,random_state=101)
```

```python
# Building Model
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
# Model Accuracy
# Method 1
logreg.score(x_test,y_test)
```

```python
# Method 2 
logreg.predict(x_test)
```

```python
print(list(y_test))
```

```python
accuracy_score(y_test,logreg.predict(x_test))
```

```python
# Building Model Using the Best 
logit_best = LogisticRegression()
logit_best.fit(x_train2,y_train2)
```

```python
# Best Accuracy
# Method 1
logit_best.score(x_test2,y_test2)
```

```python
### Single Prediction
x_test.iloc[1]
```

###SVM CLASSIFIER 

```python
clf_best = svm.SVC()
clf_best.fit(x_train2,y_train2)
```

```python
clf_best.score(x_test2,y_test2)
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train2,y_train2)
```

```python
# Model Accuracy Score
knn.score(x_test2,y_test2)
```

```python
#RFC
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(x_train2,y_train2)
pred_rfc = rfc.predict(x_test2)
```

```python
rfc.score(x_test2,y_test2)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
#Model Performance 
print(classification_report(y_test2, pred_rfc))
print(confusion_matrix(y_test2, pred_rfc))
```
